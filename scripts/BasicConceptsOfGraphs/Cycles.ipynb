```python
def draw_complete_bipartite(k: int, ax: plt.Axes):
    
    G = nx.complete_bipartite_graph(k, k)
    Y = list(G.nodes)[k:]

    # node positions
    pos = nx.bipartite_layout(
        G, Y,
        align="horizontal"
    )

    # draw nodes
    nx.draw_networkx_nodes(
        G, pos, 
        ax=ax,
        node_color="lightblue"
    )

    # node labels
    nx.draw_networkx_labels(
        G, pos,
        ax=ax,
        labels={n: n+1 for n in G.nodes}
    )
            
    # draw edges
    nx.draw_networkx_edges(
        G, pos, 
        ax=ax,
        edgelist=G.edges
    )

    ax.axis("off")
```
