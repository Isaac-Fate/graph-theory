```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_complete_bipartite(k: int, ax: plt.Axes):
    
    G = nx.complete_bipartite_graph(k, k)
    Y = list(G.nodes)[k:]

    # node positions
    pos = nx.bipartite_layout(
        G, Y,
        align="horizontal"
    )

    # draw nodes
    nx.draw_networkx_nodes(
        G, pos, 
        ax=ax,
        node_color="lightblue"
    )

    # node labels
    nx.draw_networkx_labels(
        G, pos,
        ax=ax,
        labels={n: n+1 for n in G.nodes}
    )
            
    # draw edges
    nx.draw_networkx_edges(
        G, pos, 
        ax=ax,
        edgelist=G.edges
    )

    ax.set_frame_on(False)
```

```python
# figure config
fig, axs = plt.subplots(2, 2, figsize=(16, 16), dpi=300)
axs = axs.flatten()

draw_complete_bipartite(2, ax=axs[0])
draw_complete_bipartite(3, ax=axs[1])
draw_complete_bipartite(5, ax=axs[2])
draw_complete_bipartite(10, ax=axs[3])

for i in range(4):
    axs[i].set_xlabel(f"({chr(ord('a') + i)})")
    
plt.savefig(
    "../../tex/figures/g-003.png", 
    bbox_inches="tight",
    pad_inches=0
)
```
