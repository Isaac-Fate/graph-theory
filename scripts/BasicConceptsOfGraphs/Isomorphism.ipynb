```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## Complete Bipartite Graphs

```python
m, n = 5, 3
G = nx.complete_bipartite_graph(range(1,m+1), range(m+1,m+n+1))
X = list(G.nodes)[:m]
Y = list(G.nodes)[m:]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4), dpi=150)
pos = nx.bipartite_layout(
    G, 
    nodes=Y,
    align="horizontal"
)
nx.draw_networkx_nodes(
    G, pos,
    ax=axs[0],
    node_color="lightblue"
)
nx.draw_networkx_labels(
    G, pos,
    ax=axs[0]
)
edges = list(G.edges)
nx.draw_networkx_edges(
    G, pos,
    ax=axs[0],
    edgelist=edges
)
axs[0].set_xlabel("(a)")
axs[0].set_frame_on(False)

pos = nx.circular_layout(G)
nx.draw_networkx_nodes(
    G, pos,
    ax=axs[1],
    node_color="lightblue"
)
nx.draw_networkx_labels(
    G, pos,
    ax=axs[1],
    labels={n : chr(ord("😀")+n-1) for n in G}
)
edges = list(G.edges)
nx.draw_networkx_edges(
    G, pos,
    ax=axs[1],
    edgelist=edges
)
axs[1].set_xlabel("(b)")
axs[1].set_frame_on(False)

plt.show()
```
