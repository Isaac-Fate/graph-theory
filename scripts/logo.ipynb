```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
m, n = 5, 3
G = nx.complete_bipartite_graph(range(1,m+1), range(m+1,m+n+1))
X = list(G.nodes)[:m]
Y = list(G.nodes)[m:]
```

```python
fig, ax = plt.subplots(figsize=(4, 4), dpi=150)
fig.patch.set_facecolor('black')

pos = nx.circular_layout(G)
nx.draw_networkx_nodes(
    G, pos,
    ax=ax,
    node_color="white"
)

edges = list(G.edges)
nx.draw_networkx_edges(
    G, pos,
    ax=ax,
    edgelist=edges,
    edge_color="white"
)


ax.axis("off")

plt.show()
```
